```python
import pandas as pd
import numpy as np
```

```python
X_train = np.load("data/X_train.npy",allow_pickle=True)
X_test = np.load("data/X_test.npy",allow_pickle=True)
y_train = np.load("data/Y_train.npy",allow_pickle=True)
y_test = np.load("data/Y_test.npy",allow_pickle=True)

#x_train = pd.DataFrame("data/X_train.npy")
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
df_x = pd.DataFrame(X_train[0]).rename(columns = {0: "acceleration_x", 1: "acceleration_y", 2:"acceleration_z"})
df_y = pd.DataFrame(y_train)
df_x.to_csv("data/test_folder/test.csv")
i = 1
print(f"{i}.csv")
```

```python
#Split training data by class and put it in test folder
for i in range(len(X_train)):
    x = X_train[i]
    y = y_train[i]
    df = pd.DataFrame(x).rename(columns = {0: "acceleration_x", 1: "acceleration_y", 2:"acceleration_z"})
    
    if (y == 0):
        df.to_csv(f"data/train/Under 0.08/{i}.csv")
        print(f"index {i} saved to under 0.08 file")
    elif (y == 1): 
        df.to_csv(f"data/train/Over 0.08/{i}.csv")
        print(f"index {i} saved to over 0.08 file")

    else:
        print("data point did not have proper y value and was not saved to a folder")
```

```python
#Do the same for test data
for i in range(len(X_test)):
    x = X_test[i]
    y = y_test[i]
    df = pd.DataFrame(x).rename(columns = {0: "acceleration_x", 1: "acceleration_y", 2:"acceleration_z"})
    
    if (y == 0):
        df.to_csv(f"data/test/Under 0.08/{i}.csv")
        print(f"index {i} saved to under 0.08 file")
    elif (y == 1): 
        df.to_csv(f"data/test/Over 0.08/{i}.csv")
        print(f"index {i} saved to over 0.08 file")

    else:
        print("data point did not have proper y value and was not saved to a folder")
```
